```python
# Function of this document is for me to have an outline for making animations
# "Stolen" from Maja
from tqdm import tqdm
import matplotlib.pyplot as plt
import numpy as np
from perlin_noise import PerlinNoise
import matplotlib.animation as animation
from IPython.display import HTML

# Setting up a plot space
fig, ax = plt.subplots()
ax.axis('equal')            # Forcing it to be equal in axis ratios

x_min = 0
x_max = 40
y_min = 0
y_max = 40

ax.set_xlim(x_min, x_max)
ax.set_ylim(y_min, y_max)

def get_data(octave, x_max):
    """Animator requires any function for generating data"""
    noise1 = PerlinNoise(octaves=octave)
    noise2 = PerlinNoise(octaves=octave*2)
    noise3 = PerlinNoise(octaves=octave*4)
    xpix, ypix = x_max, x_max
    
    pic = []
    for i in range(xpix):
        row = []
        for j in range(xpix):
            noiseval = 0
            noiseval =+ noise1([i/xpix, j/ypix])
            noiseval =+ (1/2) * noise2([i/xpix, j/ypix])
            noiseval =+ (1/4) * noise3([i/xpix, j/ypix])
            row.append(noiseval)
        pic.append(row)
    return pic

def update(i):
    """
    Update function is what is happening every frame. Imagine that you have to make a new plot every frame,
    but instead you clear the background and overlay something else.
    """
    # Step 1 # Generate data
    pic = get_data(4, x_max)

    # Step 2 # Clear previous plot and plot the updated plot
    ax.clear()          # Alternatively ax.clear() Clears the entire thing, but collections offers selection
    ax.imshow(pic, cmap="gray")

    # Step 3 # Modification of axis layout should occur in here
    ax.set_xlim(0, x_max)
    ax.set_ylim(0, y_max)
    
    return None


anim = animation.FuncAnimation(fig, update, frames=100, interval=250) # Controls the plotted animation
plt.rcParams['animation.embed_limit'] = 2**128          # Limits the Bytes allocated to the animation
anim.save('Test_animation_Perlin.gif', writer='imagemagick', fps=4)      # Controls the animation for the downloadable gif
plt.close(fig)

HTML(anim.to_jshtml())      # Converts anim to an actual gif
```
