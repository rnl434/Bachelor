```python
import matplotlib.pyplot as plt 
import numpy as np 
import random as r 
from scipy.spatial import Voronoi, voronoi_plot_2d
```

```python
# Setting seed #
seed = None
r.seed(seed)
np.random.seed(seed)
### Random Sample ###
sample_size = 5
data_x = np.random.rand(sample_size)
data_y = np.random.rand(sample_size)
points = []
for i in range(len(data_x)):
    point = [data_x[i], data_y[i]]
    points.append(point)


### Plotting ###
vor = Voronoi(points)
voronoi_plot_2d(vor, show_vertices = False);

plt.grid();
plt.xlim(0, 1)
plt.ylim(0, 1);
```
